```python
from tkinter import *
from PIL import Image, ImageTk
import os
import numpy as np
import tensorflow as tf


modelFullPath = 'output_graph3.pb'
labelsFullPath = 'output_labels3.txt'

path='myimage.jpg'


def create_graph():
    """Creates a graph from saved GraphDef file and returns a saver."""
    # Creates graph from saved graph_def.pb.
    with tf.gfile.FastGFile(modelFullPath, 'rb') as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())
        _ = tf.import_graph_def(graph_def, name='')
        
        
        
def run_inference_on_image2(path):
    answer = None
    image_data = tf.gfile.FastGFile(path, 'rb').read()

    create_graph()

    with tf.Session() as sess:

        softmax_tensor = sess.graph.get_tensor_by_name('final_result:0')
        predictions = sess.run(softmax_tensor,
                               {'DecodeJpeg/contents:0': image_data})
        predictions = np.squeeze(predictions)

        top_k = predictions.argsort()[-5:][::-1]  # Getting top 5 predictions
        f = open(labelsFullPath, 'rb')
        lines = f.readlines()
        labels = [str(w).replace("\n", "") for w in lines]
       
        return labels, predictions

def set_text(path):
    
    labels=['Blackbird Origin:  American oscine birds ', 'Starling starling Origin: New Zealand', 'Mesange Bleue  Origin:Europe and western Asia', 'Blue Tit Origin:UK', 'Robin Origin:', 'Message Charbonniere Origin:North Africa', 'House sparrows Origin:North Africa', 'pouillot fitis Origin:French ', 'Fast pouillot Origin:Old French word poulet']
    a, b=run_inference_on_image2(path) 
    
    indices = np.where(b == b.max())
    x=indices[0]
    x=x[0]
    text=labels[x]
    e.config(text = text, fg="dark green", font=(None, 20))
    
    return
    
def pop_window():
    import easygui
    from PIL import Image, ImageTk
    path = easygui.fileopenbox(default="exemple_pictures/*.jpg", filetypes=['exemple_pictures/*.jpg'])
    print(path)
    img2 = Image.open(path)
    img2=img2.resize((300,250),Image.ANTIALIAS)
    img2.save('myimage.jpg','jpeg')
    #print(type(img2))
    img2=ImageTk.PhotoImage(img2)

    panel.configure(image = img2)
    panel.image = img2
    e.config(text = '')

root = Tk()
root.geometry("800x600")
root.resizable(0, 0) #

topbar=Label(root, bg="#009688", fg="white", height=3, width=1500, text="Bird Recognition", font=("Helvetica", 20))
topbar.pack()


im1=Image.open(path)
img=im1.resize((300,250),Image.ANTIALIAS)
img3=ImageTk.PhotoImage(img)
panel = Label(root, image = img3)

panel.place(relx=.5, rely=.4, anchor="c")

bouton_load = Button(root, text="Load image", command=lambda:pop_window())
bouton_load.place(relx=.35, rely=.65, anchor="c")


bouton_analyser = Button(root, text="Run Diagnosis", command=lambda:set_text(path))
bouton_analyser.place(relx=.65, rely=.65, anchor="c")

e=Label(root, text = "", font=(None, 15), width=200, )
e.place(relx=.5, rely=.85, anchor="c")

bouton_quitter = Button(root, text="Quit", command=root.quit)
bouton_quitter.place(relx=.9, rely=.9, anchor="c")


res = Label(root)
res.pack()
root.mainloop()
```
